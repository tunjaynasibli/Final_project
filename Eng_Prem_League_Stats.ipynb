### Max Odds are only available for season19 and season20, so to find out most underrated team by bookmakers I will use these 2 files.

```python
seasonodds=pd.concat([season19, season20], axis=0)
```

```python
# Below we use function to find how profitable to bet on each result. HomeTeam win, draw or AwayTeam win
```

```python
seasonodds['HomeOdds'] = seasonodds.apply(lambda x: x['MaxH'] - 1 if x['FTR'] == 'H' else -1, axis=1)
```

```python
seasonodds['AwayOdds'] = seasonodds.apply(lambda x: x['MaxA'] - 1 if x['FTR'] == 'A' else -1, axis=1)
```

```python
seasonodds["DrawOdds"] = seasonodds.apply(lambda x: x['MaxD'] - 1 if x['FTR'] == 'D' else -1, axis=1)
```

```python
### by grouping we will find out total result of 2 seasons performance. 
```

```python
homeodd=seasonodds.groupby(["HomeTeam"])["HomeOdds"].sum()
```

```python
awayodd=seasonodds.groupby(["AwayTeam"])["AwayOdds"].sum()
```

```python
totalodds=homeodd + awayodd
underrated=totalodds.sort_values(ascending=False).head(3)
```

```python
data_pred
```

```python
# generating mean of variable until last game 
def avg_season(var):

    media = []
    for i, j in zip(data_pred["HomeTeam"],data_pred["Date"]):
        oi = data_pred.loc[data_pred["HomeTeam"]<i]
        oi = oi.loc[oi["Date"]==j]
        oi = oi[var].mean()
        media.append(oi)

    data_pred["avrg_"+var] = pd.DataFrame(np.array(media))
    

# generating mean of variable for the last 3 game 
def avg_last3_season(var):

    media = []
    for i, j in zip(data_pred["HomeTeam"],data_pred["Date"]):
        oi = data_pred.loc[data_pred["HomeTeam"]<i]
        oi = oi.loc[oi["Date"]==j]
        oi= oi.reset_index(drop=True)
        oi = oi[-3:]
        oi = oi[var].mean()
        media.append(oi)

    data_pred["last3_avg_"+var] = pd.DataFrame(np.array(media))
```

```python
#using all created functions 

var_avg = data_pred.drop(columns=['Date','HomeTeam','AwayTeam', 'HTR', 'FTR'])

for i in var_avg.columns:
    avg_season(i)
    avg_last3_season(i)
    
    
```
