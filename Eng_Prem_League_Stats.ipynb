```python
import pandas as pd
```

```python
# uploading csv files, each file is single season
season20 = pd.read_csv("E1.csv")
season19 = pd.read_csv("E 19:20.csv")
season18 = pd.read_csv("E 18:19.csv")
season17 = pd.read_csv("E 17:18.csv")
season16 = pd.read_csv("E 16:17.csv")
```

```python
# concatinating seasons in a single data
data=pd.concat([season16, season17, season18,season19,season20], axis=0)
```

```python
data["Home_points"]=3
data["Away_points"]=3
```

```python
data= data.reset_index(drop=True)
```

```python
for index, row in data["FTR"].items():   
    
    if row=="A":
        data["Away_points"][index]=3
        data["Home_points"][index]=0
    elif row=="D":
        data["Away_points"][index]=1
        data["Home_points"][index]=1
    elif row=="H":
        data["Away_points"][index]=0
        data["Home_points"][index]=3

        
```

```python
data
```

```python
home=data.groupby(["HomeTeam"])["Home_points"].sum()
```

```python
away=data.groupby(["AwayTeam"])["Away_points"].sum()
```

```python
total=home+away
```

```python
total
```

```python
homecount=data.groupby(["HomeTeam"])["Home_points"].count()
```

```python
awaycount=data.groupby(["AwayTeam"])["Away_points"].count()
```

```python
total_played=homecount+awaycount
total_played
```

```python
ppg=total/total_played
```

```python
ppg
```

```python
ppg.to_csv("ppg.csv")
```

```python
season16stats = pd.read_csv("pl_16-17.csv")
season17stats = pd.read_csv("pl_17-18.csv")
season18stats = pd.read_csv("pl_18-19.csv")
season19stats = pd.read_csv("pl_19-20.csv")
season20stats = pd.read_csv("pl_20-21.csv")
```

```python
dataset=pd.concat([season16stats, season17stats, season18stats,season19stats,season20stats], axis=0)
```

```python
dataset=dataset.reset_index()
```

```python
dataset.columns
```

```python
dataset['Successful 50/50s']
```

```python
rem_list=['index', 'Unnamed: 0', 'Clean sheets', 'Goals conceded', 'Last man tackles', "Blocked shots", 
          'Headed Clearance', 'Clearances off line', 'Recoveries', 'Successful 50/50s', 'Duels lost', 
          'Headed goals','Aerial battles lost', 'Own goals', 'Through balls', 'Yellow cards', 'Red cards', 'Fouls',
          'Offsides','Goals with right foot', 'Goals with left foot','Hit woodwork', 'Goals per match','Throw outs',
          'Goal Kicks','Shooting accuracy %','Big chances missed','Punches','High Claims','Catches',
          'Sweeper clearances','Big Chances Created','Penalties Saved', "Interceptions", "Clearances"
         ]
```

```python
dataset.drop(rem_list, axis=1, inplace=True)
```

```python
dataset
```

```python
Tackles=dataset.groupby(["Name"])["Tackles"].sum()
```

```python
Tackles.sort_values(ascending=False).head(10)
```

```python
Appearances=dataset.groupby(["Name"])["Appearances"].sum()
Appearances.sort_values(ascending=False).head(10)
```

```python
tackles_pg=Tackles/Appearances
tackles_pg.sort_values(ascending=False).head(10)
```

```python
Duels=dataset.groupby(["Name"])["Duels won"].sum()
Duels.sort_values(ascending=False).head(10)
```

```python
Duels_pg=Duels/Appearances
Duels_pg.sort_values(ascending=False).head(10)
```

```python
Assists=dataset.groupby(["Name"])["Assists"].sum()
assists_top5=Assists.sort_values(ascending=False).head(5)
```

```python
assists_top5.to_csv("assists.csv")
```

```python
dataset["Passes"] = dataset["Passes"].str.replace(r',', '')
```

```python
dataset["Passes"] = dataset["Passes"].astype(str).astype(int)
```

```python
Passes=dataset.groupby(["Name"])["Passes"].sum()
passes_pg=Passes/Appearances
passes_pg=passes_pg.sort_values(ascending=False).head(5)
```

```python
passes_pg.to_csv("passes.csv")
```
